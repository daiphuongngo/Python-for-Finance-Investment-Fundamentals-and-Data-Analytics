```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 170
rev_stdev = 20
iterations = 1000
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
plt.figure(figsize=(15, 6))
plt.plot(rev)
plt.show()
```

```python
COGS = - (rev * np.random.normal(0.6,0.1))
 
plt.figure(figsize=(15, 6))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
# The difference between revenues and Coggs gives those Gross Profit 
Gross_Profit = rev + COGS
Gross_Profit

# It is normally distributed with mean value equal to the difference between the mean of the revenues and the mean of COGS
plt.figure(figsize=(15, 6))
plt.plot(Gross_Profit)
plt.show()
```

```python
max(Gross_Profit)
```

```python
min(Gross_Profit)
```

```python
Gross_Profit.mean()
```

```python
Gross_Profit.std()
```

```python
plt.figure(figsize=(10, 6));
plt.hist(Gross_Profit, bins = [40, 50, 60, 70, 80, 90, 100, 110, 120]); # 1st way: create a list whose elements separate the bins along the x axis
plt.show()
```

```python
plt.figure(figsize=(10, 6));
plt.hist(Gross_Profit, bins = 20); # 2nd way: assign the number of bins directly
plt.show()
# It will be obvious the numbers are normally distributed.
# It helps to predict future revenues costs and margins.
```
