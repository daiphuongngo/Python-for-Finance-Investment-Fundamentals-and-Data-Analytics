```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['^GSPC', '^IXIC', '^GDAXI']

ind_data = pd.DataFrame()

for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source='yahoo', start='1997-1-1')['Adj Close']
```

```python
ind_data.head()
```

```python
ind_data.tail()
```

```python
# Normalize to 100
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(15, 6));
plt.show()

# We observe a general rise of the indices value.
# This was because in the period between 1995 and 2001 there was a dot.com bubble and the stock markets in industrialized 
#countries like USA UK and Germany were boosted by a significant rise in the value of Internet companies being composed 
#primarily of tech companies Nasdaq logically depicts best this trend relative to the other indices in a certain period its 
#performance declines abruptly in the next few years.
# Germany manifest stability and growth.
# Of course the global economic and financial crises of 2007 2008 as well as the sovereign debt crisis in 2011 reflect 
#the picture accordingly.
# But in the long term observing the results from 1997 until today we can say that overall all indices performed very well 
#with only an occasional drop of the FTSE below the initial growth level.
# Nasdaq scores highest in certain periods and its performance is connected to slightly sharper variations at times.
# The German market is the one that demonstrates real persistence and stability.
```

```python
# Calculate the indices simple returns
ind_returns = (ind_data / ind_data.shift(1)) - 1

ind_returns.tail()
```

```python
# Show annual returns --->>> all 4 numbers are positive
# The implication is that the average of the 4 companies listed under the four indices has provided investors
# with a positive rate of return over the past two decades.
# Apparently at least for this time span Nasdaq surpasses the S&P 500 easily the S&P 500 comes in third. 
# Actually because the German DAX outperformed it.
annual_ind_returns = ind_returns.mean() * 250
annual_ind_returns
```
