```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Goal: predict the firm's future gross profit
# Requirements:
# - Expected revenue
# - Expected Cost of Goods Sold (COGS)
# Steps:
# - Performing 1000 simulations of the company's expected revenues
# - Expected revenues for this year are $170 billion.
# - Standard deviation of this figure would be $20 billion.
# - To simplify things let's work in millions of dollars.
```

```python
rev_m = 170
rev_stdev = 20
iterations = 1000
```

```python
# Apply numpy random normal distribution generator
rev = np.random.normal(rev_m, rev_stdev, iterations) 
rev
```

```python
plt.figure(figsize=(15, 6))
plt.plot(rev)
plt.show()
```

```python
# The simulated values are centered on the mean of 170.
# It is obvious that almost all data points fall in the region of 150 to 190.
# The two values equal the mean minus one standard deviation and the mean plus 1 standard deviation for 1000 data points. 
# We rarely observe scenarios outside of these limits.
```

```python
# Assume COGS ~ approx 60% of the revenues & standard deviation of 10% -->> spent to produce the goods it sells
COGS = - (rev * np.random.normal(0.6,0.1)) # COGS are money spent -> minus
# Reflect the multiplication of revenues by 60%

plt.figure(figsize=(15, 6))
plt.plot(COGS)
plt.show()
```

```python
# We will not simulate COGS 1000 times.
# This has already been done for revenues and we have 1000 revenue data points.
# We must assign a random COGS value to each one of these points.
# Coggs is a percentage of revenues.
# This is why the revenue value we obtained must be multiplied by a value extracted from a random normal distribution with a mean of 
# 0.6 and a standard deviation of 0.1.
# Had we put 0.6 directly that would have meant Coggs always equals 60 percent of revenues. And this is not the case.
# Percentage will probably vary and we have decided that standard deviation is equal to 10%.
# Numpy allows us to incorporate the expected deviation.
# What is important is that every time the deviation of cogs will be around 10% of its mean.
```
